```python
cf.insert(20, '2021-2022 Segment Enrollment', 'x')
```

```python
for row in cf.iterrows():

    # Get the value of the Borough column
    borough = row[1]['Borough']

    # Get the value of the Public Status column
    public_status = row[1]['Public Status']

    # Check if the Borough column contains Brooklyn
    if borough == 'Brooklyn':

        # Check if the Public Status column contains Public
        if public_status == 'Public':

            # Write 513412 in the 2021-2022 Segment Enrollment column
            cf.loc[row[0], '2021-2022 Segment Enrollment'] = 260855

        else:

            # Write 116725 in the 2021-2022 Segment Enrollment column
            cf.loc[row[0], '2021-2022 Segment Enrollment'] = 116725
    elif borough == 'Bronx':
         # Check if the Public Status column contains Public
        if public_status == 'Public':

            # Write 513412 in the 2021-2022 Segment Enrollment column
            cf.loc[row[0], '2021-2022 Segment Enrollment'] = 168582

        else:

            # Write 116725 in the 2021-2022 Segment Enrollment column
            cf.loc[row[0], '2021-2022 Segment Enrollment'] = 23954
            
    elif borough == 'Queens':
         # Check if the Public Status column contains Public
        if public_status == 'Public':

            # Write 513412 in the 2021-2022 Segment Enrollment column
            cf.loc[row[0], '2021-2022 Segment Enrollment'] = 263898
        else:

            # Write 116725 in the 2021-2022 Segment Enrollment column
            cf.loc[row[0], '2021-2022 Segment Enrollment'] = 39686
    elif borough == 'Manhattan':
         # Check if the Public Status column contains Public
        if public_status == 'Public':

            # Write 513412 in the 2021-2022 Segment Enrollment column
            cf.loc[row[0], '2021-2022 Segment Enrollment'] = 152463
        else:

            # Write 116725 in the 2021-2022 Segment Enrollment column
            cf.loc[row[0], '2021-2022 Segment Enrollment'] = 49856
    else:
        if public_status == 'Public':

            # Write 513412 in the 2021-2022 Segment Enrollment column
            cf.loc[row[0], '2021-2022 Segment Enrollment'] = 61797
        else:

            # Write 116725 in the 2021-2022 Segment Enrollment column
            cf.loc[row[0], '2021-2022 Segment Enrollment'] = 11337
```
